Problem 1:

```python
def func(x):
    return (math.log(x/2)-math.sin(5*x/2))

def d_func(x):
    return ((1/x)-math.cos(5*x/2)*2.5)

a0, b0 = lb.bracketing(1.5, 2.5, func) # bracketing interval

print("[",a0,",",b0,"]")

print("The convergence for Regula Falsi:")
lb.Rf(func,a0,b0)

print("The convergence for Newton Raphson:")
lb.Nr(func,d_func,0.5)
```

PROBLEM 2: HEAT EQUATION

```python
#the variables
x0=0
xf=10
y0=40
yf=200
z0=1
dx=0.01
tol=0.001

def T2(x, T, t):
    return -1*0.01*(20-T)

def T1(x, T, t):
    return t

#using shooting method
x, T, t = lb.Shoot(T2, T1, x0, xf, y0, yf, z0, dx, tol, 'yes')

#finding closest temp to 100 in the rod
print('\n The temp at which rod is 100 +- 0.1 C is:\n')
for i in range(0,len(x)):
    if abs(T[i]-100) <= 0.1:
        print(T[i], 'C at', x[i],'m.')
```

Problem 3: heat equation

```python
time_step=0.001
x_step=0.1
x_limit=2
total_time=1 #proportional to number of iterations
nx=round(x_limit/x_step) #number of steps in x  

u=[0]*nx
u[round(nx/2)]=300 #intial temp profile

u=lb.epde(u,time_step,x_step,x_limit,total_time)
plt.plot(u)
plt.xlabel("x")
plt.ylabel("temp")
plt.title("temp profile after 1000 timesteps")
plt.show()
```

PROBLEM: 4  centre of mass 

```python
def f1(x):
    return x**2

def f2(x):
    return x**3

mx=lb.com_i(0,2,f2,50)
Total_mass=lb.com_i(0,2,f1,50)
com=mx/Total_mass

print("The rod is placed along x axis with one end at origin and other end at x= 2 metres")
print("Centre of mass coordinate  is",round(com,4),"metres")
```

Problem 5: lu decomposition

```python
data=[]#Reading the input file
with open("input1.txt") as f:
    for line in f:
        data.append(line.split(","))
n=len(data)       
a=np.zeros((n,n))#matrix
b=np.zeros(n)#vector
for i in range(n):
    b[i]=(int)(data[i][n])
    for j in range(n):
        a[i][j]=(int)(data[i][j])

lb.LU_decompose(a,b)  
print("The solution of the linear equation using lu decomposition is : ", b)
```
